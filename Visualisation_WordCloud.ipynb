```python
import numpy as np
import random
from os import path
from PIL import Image
import PIL
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (16.0, 9.0)
```

```python
df_book1 = pd.read_csv('AGOT.csv')
df_book2 = pd.read_csv('ACOK.csv')
df_book3 = pd.read_csv('ASOS.csv')
df_book4 = pd.read_csv('AFFC.csv')
df_book5 = pd.read_csv('ADWD.csv')
```

```python
dfStr_book1 = df_book1.to_string()
dfStr_book2 = df_book2.to_string()
dfStr_book3 = df_book3.to_string()
dfStr_book4 = df_book4.to_string()
dfStr_book5 = df_book5.to_string()
```

```python
masked_stark = np.array(Image.open(r"C:\Users\Kanishk Verma\Desktop\Course\CA682\TowerOfHand Data\stark.jpg"))
masked_baratheon = np.array(Image.open(r"C:\Users\Kanishk Verma\Desktop\Course\CA682\TowerOfHand Data\tyrion.png"))
masked_lannister = np.array(Image.open(r"C:\Users\Kanishk Verma\Desktop\Course\CA682\TowerOfHand Data\lannister.jpg"))
masked_targaryen = np.array(Image.open(r"C:\Users\Kanishk Verma\Desktop\Course\CA682\TowerOfHand Data\targaryen.jpg"))
masked_cersei = np.array(Image.open(r"C:\Users\Kanishk Verma\Desktop\Course\CA682\TowerOfHand Data\cersei-edit.png"))
```

```python
add_stopwords = ['The', 'He', 'She', 'will', 'Says', 'says', 'asks', 'go', 'take' ,'goes', 'want', 'Lord', 'Ser', 'lord',
                'way', 'ways', 'Way', 'Ways', 'want', 'Wants', 'wants', 'Want', 'tells', 'one', 'man', 'men', 'think',
                'thinks', 'three', 'will', 'must', 'They', 'three', 'well', 'time', 'back', 'As', 'two', 'night', 'When',
                'day', 'next', 'still', 'see', 'Maester', 'comes', 'one', 'first', 'killed', 'now', 'make', 'leave', 'A',
                'Lady', 'lady', 'years', 'takes', 'castle', 'son', 'new', 'come']
```

```python
stopwords = set(STOPWORDS)
for words in add_stopwords:
    stopwords.add(words)
```

```python
def grey_color_func(word, font_size, position, orientation, random_state=None,
                    **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(60, 100)
```

```python
wc1 = WordCloud(background_color="black", stopwords = stopwords, width=400, height=400, contour_width=0.5,
                contour_color='white', mask=masked_stark, random_state=3)
wc2 = WordCloud(background_color="black", stopwords = stopwords, width=400, height=400, contour_width=0.5,
                contour_color='white', mask=masked_baratheon, random_state=3)
wc3 = WordCloud(background_color="black", stopwords = stopwords, width=400, height=400, contour_width=0.5,
                contour_color='white', mask=masked_lannister, random_state=3)
wc4 = WordCloud(background_color="black", stopwords = stopwords, width=400, height=400, contour_width=0.5,
                contour_color='white', mask=masked_targaryen, random_state=3)
wc5 = WordCloud(background_color="black", stopwords = stopwords, width=400, height=400, contour_width=0.5,
                contour_color='white', mask=masked_cersei, random_state=3)
```

```python
wc1.generate(dfStr_book1)
wc2.generate(dfStr_book2)
wc3.generate(dfStr_book3)
wc4.generate(dfStr_book4)
wc5.generate(dfStr_book5)
```

```python
plt.figure(figsize=[20,10])
plt.imshow(wc5.recolor(color_func=grey_color_func, random_state=3),
           interpolation="bilinear")
plt.axis("off")
plt.show()
```
