```python
import re
def cleanFile(Text):
    cleanText = []
    for i in Text:
        wor = i.replace("'s", "")
        wor = wor.replace("\n", "")
        wor = wor.replace('!',"")
        wor = wor.replace(',',"")
        wor = wor.replace('.',"")
        wor = wor.replace('?',"")
        wor = wor.replace('[',"")
        wor = wor.replace(']',"")
        wor = wor.replace(':',"")
        cleanText.append(wor)
        return cleanText
```
